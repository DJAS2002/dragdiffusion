## Load the metrics and filter results

We start of with the L1, L2 experiments

```python
import pandas as pd

l1l2_metrics=pd.read_csv('evaluation_results_L1L2.csv')
l1l2_metrics
```

Change the eval_root strings into something more readable focusing on the L1 and L2 combinations

```python
import re

# Function to process the string
def process_string(string):
    # Find all occurrences of L1m, L1p, L1mask with their values
    matches = re.findall(r'(L1\w+)=([^_]+)', string)
    result = []
    for key, value in matches:
        # If the value is False, replace L1 with L2
        if value == 'False':
            result.append(key.replace('L1', 'L2'))
        else:
            result.append(key)
    return ' '.join(result)

# Apply the function to the column
l1l2_metrics['Eval_Root'] = l1l2_metrics['Eval_Root'].apply(process_string)

print(l1l2_metrics)
```

### Pivot the data to create an overview for each category and save it to csv.

CLIP similarity overview

```python
l1l2_metrics_pivot = l1l2_metrics.pivot_table(
    index=["Category", "Metric"],  # Keep Category and Metric as indices
    values="Value",               # Use 'Value' as the data
    columns="Eval_Root"           # Turn 'Eval_Root' into columns
).reset_index()

l1l2_metrics_pivot_CLIP= l1l2_metrics_pivot[l1l2_metrics_pivot['Metric']=='avg CLIP similarity']
l1l2_metrics_pivot_CLIP.to_csv('evaluation_grid_CLIP.csv', index=False)
l1l2_metrics_pivot_CLIP
```

LPIPS overview

```python
l1l2_metrics_pivot_LPIPS= l1l2_metrics_pivot[l1l2_metrics_pivot['Metric']=='avg LPIPS']
l1l2_metrics_pivot_LPIPS.to_csv('evaluation_grid_LPIPS.csv', index=False)
l1l2_metrics_pivot_LPIPS
```

Mean Distance (MD) overview

```python
l1l2_metrics_pivot_MD= l1l2_metrics_pivot[l1l2_metrics_pivot['Metric']=='avg MD']
l1l2_metrics_pivot_MD.to_csv('evaluation_grid_MD.csv', index=False)
l1l2_metrics_pivot_MD
```
