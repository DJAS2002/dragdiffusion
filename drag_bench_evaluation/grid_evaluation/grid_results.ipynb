```python
import pandas as pd

grid_metrics=pd.read_csv('evaluation_results.csv')
grid_metrics
```

```python
import re

# Function to process the string
def process_string(string):
    # Find all occurrences of L1m, L1p, L1mask with their values
    matches = re.findall(r'(L1\w+)=([^_]+)', string)
    result = []
    for key, value in matches:
        # If the value is False, replace L1 with L2
        if value == 'False':
            result.append(key.replace('L1', 'L2'))
        else:
            result.append(key)
    return ' '.join(result)

# Apply the function to the column
grid_metrics['Eval_Root'] = grid_metrics['Eval_Root'].apply(process_string)

print(grid_metrics)
```

```python
grid_metrics_pivot = grid_metrics.pivot_table(
    index=["Category", "Metric"],  # Keep Category and Metric as indices
    values="Value",               # Use 'Value' as the data
    columns="Eval_Root"           # Turn 'Eval_Root' into columns
).reset_index()

grid_metrics_pivot_CLIP= grid_metrics_pivot[grid_metrics_pivot['Metric']=='avg CLIP similarity']
grid_metrics_pivot_CLIP.to_csv('evaluation_grid_CLIP.csv', index=False)
grid_metrics_pivot_CLIP
```

```python
grid_metrics_pivot_LPIPS= grid_metrics_pivot[grid_metrics_pivot['Metric']=='avg LPIPS']
grid_metrics_pivot_LPIPS.to_csv('evaluation_grid_LPIPS.csv', index=False)
grid_metrics_pivot_LPIPS
```
